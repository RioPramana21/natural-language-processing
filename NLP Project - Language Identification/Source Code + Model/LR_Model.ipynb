```python
import pickle
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
import warnings
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn import pipeline
from sklearn.linear_model import LogisticRegression
from datasets import load_dataset
warnings.simplefilter("ignore")
```

## Load and Split Datasets

```python
#Load dataset from Hugging Face
ds_train = load_dataset('papluca/language-identification', split='train')
ds_train = ds_train.to_pandas()
ds_test = load_dataset('papluca/language-identification', split='test')
ds_test = ds_test.to_pandas()
```

```python
#Split features and target
x_train, y_train = ds_train["text"], ds_train["labels"] #train dataset
x_test, y_test = ds_test['text'], ds_test['labels'] #test dataset
```

## Preprocessing Text Functions

```python
#Preprocessing functions
def clean_text(text):
    text = text.lower()
    text = re.sub(r'(@[A-Za-z0-9]+)|([!"#$%&\'()*+,-./:;<=>?@[\]^_`{|}])|(\w+://\S+)|^rt|http.+?', '', text)
    return text

def preprocess_multiple_input(x):
  data_list = []
  for text in x:
    text = clean_text(text)
  data_list.append(text)
  return data_list
```

```python
#Encode target variable
le = LabelEncoder()
y_train = le.fit_transform(y_train)
y_test = le.transform(y_test)
```

## Make and Train Models based on 3 Variations of Vectorizer
We then save the model into a .pkl file

```python
#Fit and save the model
vectorizer = TfidfVectorizer(ngram_range=(1,3), analyzer='char')
model = pipeline.Pipeline([
    ('vectorizer', vectorizer),
    ('clf', LogisticRegression())
])
model.fit(x_train,y_train)
with open("lr_model.pkl", "wb") as f:
  pickle.dump(model, f)
```

```python
#Test countvectorizer
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
model = pipeline.Pipeline([
    ('vectorizer', vectorizer),
    ('clf', LogisticRegression())
])
model.fit(x_train,y_train)
with open("lr_model_count.pkl", "wb") as f:
  pickle.dump(model, f)
```

```python
#Test countvectorizer with params
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(ngram_range=(1,3), analyzer='char')
model = pipeline.Pipeline([
    ('vectorizer', vectorizer),
    ('clf', LogisticRegression())
])
model.fit(x_train,y_train)
with open("lr_model_param.pkl", "wb") as f:
  pickle.dump(model, f)
```
